### Math

```python
x = int(input("what's is x? "))
y = int(input("what's is y? "))
print(x+y)

#round(number[, ndigits])
x = float(input("what's is x? "))
y = float(input("what's is y? "))
z = round(x/y, 2) #just take 2 decimal after the .
print(z)

#just for the more formatting
z = x/y
print(f"{z:.2f}")
```
